## Required Sections

**1. Introduction (5%)**

Mention which problem you are working on.  Include the link to the Kaggle webpage.  Describe the 
problem as you understand it and summarize what techniques you tried.f

**2. Data Exploration (20%)**

  - Did you look at sample data to understand it better?
  - Did you visualize prior distributions of features
  - Did you characterize individual feature distributions vs. the target
  - Did you characterize features vs. each other.
  - Did you providde some exposition describing relationships as well
    as how they relate to the objective.
  - Did you note any nonlinearities or peculiarities where dimensionality
    reduction or other feature engineering might be relevant (bonus)

**3. Preprocessing (20%)**

  - did you check whether there are missing data points?
  - did you check what fraction of the training samples are so affected?
  - did you omit data points with missing data?
  - how many data points in total have at least one field that is missing?
  - did you use imputation to fill in (synthesize) missing data points?
  - did you treat the fact that there is missing data as itself
    a feature (may depend on the problem)

**4. Use of Multiple Methods / hyperparameter tuning / etc. (35%)**

**5. Validation, Summarization and Visualization of Results (20%)**

  - classification report,  confusion matrices, etc.
