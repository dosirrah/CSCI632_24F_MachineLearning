# CSCI 632: Homework 1

Machine learning makes heavy use of linear algebra and probability
theory.  This homework contains problems first to learn and practice
LaTeX.  We then move on to linear algebra.

Linear algebra is great at representing problems with high dimensionality.
In machine learning we go a little beyond what you would normally have in an
introductory linear algebra course: tensors.

Even for those that know linear algebra reasonably well, there are
some tensors, Hadamard product, and a few other concepts are often
not covered in an introductory linear algebra course.


## Notation

Example              | Meaning
---------------------|------------------------------------------------
$a$                  | scalar (integer or real number)
$\mathbf{a}$         | vector
$\mathbf{A}$         | matrix
$\mathsf{A}$         | tensor
$\mathbf{e}^{(i)}$   | standard basis vector (i.e., one-hot vector)
$\mathbf{A}^T$       | tranpose of matrix $\mathbf{A}$
$\mathbf{A}_{i,j}$   | element at $i^{th}$ row and $j^{th}$ column

## Part 2. Column Vectors

In geometry a vector has often been represented as a tuple like $(a,b)$.
When using matrices, we often represent a vector as a column vector.

**Problem 2.1**

In Python, plot the following using matplotlib and its `quiver` function.  The axes
should pass through the origin, and the axes should be labelled as $x$ and $y$.

$\begin{bmatrix}
7 \\
3
\end{bmatrix}
$

Your plot should look similar to this:

<!---->




**Answer 2.1.**

It is not necessary for the plot to look exactly like this to get full credit, but it should
at least show a vector starting from an origin and reaching the coordinate $(7,3)$.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

ax.quiver(0, 0, 7, 3, scale=1, scale_units='xy', angles='xy', color='r')

# Setting the limits and aspect ratio
ax.set_xlim(-1, 8)
ax.set_ylim(-1, 4)
ax.set_aspect('equal')

# Adding grid
ax.grid(True)

# Remove the heavy line along the top and bottom of the graph.
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Move the axes so that they extend from the origin.
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# Add x and y labels
ax.text(8.1, 0, 'x')  # 8.1 is in data units.
ax.text(0, 4.1, 'y')



# Show the plot
plt.show()
```
