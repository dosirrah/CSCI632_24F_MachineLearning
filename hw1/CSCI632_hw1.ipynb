    def numpy_to_latex_matrix(matrix):
        latex_str = r'\begin{bmatrix}'

**An Acceptable Answer to Example 6.A**

The following is an appropriate answer.

$

\begin{bmatrix}
 1 & 2 \\
 3 & 4

\begin{bmatrix}
 5 & 6 \\
 7 & 8

 3 \cdot 5 + 4 \cdot 7 & 3 \cdot 6 + 4 \cdot 8
